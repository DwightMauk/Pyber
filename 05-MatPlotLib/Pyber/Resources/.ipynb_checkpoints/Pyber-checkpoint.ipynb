```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

city_data = '../Resources/city_data.csv'
ride_data = '../Resources/ride_data.csv'

city_df = pd.read_csv(city_data)
ride_df = pd.read_csv(ride_data)

#city_df.head()
#ride_df.head()
city_df = city_df.drop_duplicates('city', keep = 'first')
pyber_data = pd.merge(ride_df,city_df,on='city',how='left')
pyber_data.head()
```

```python
# Obtain the x and y coordinates for each of the three city types
urban_city = pyber_data[pyber_data["type"] == "Urban"]
sub_city = pyber_data[pyber_data["type"] == "Suburban"]
rural_city = pyber_data[pyber_data["type"] == "Rural"]

urban_ride_count = urban_city.groupby(["city"]).count()["ride_id"]
urban_avg_fare = urban_city.groupby(["city"]).mean()["fare"]
urban_driver_count = urban_city.groupby(["city"]).mean()["driver_count"]

sub_ride_count = sub_city.groupby(["city"]).count()["ride_id"]
sub_avg_fare = sub_city.groupby(["city"]).mean()["fare"]
sub_driver_count = sub_city.groupby(["city"]).mean()["driver_count"]

rural_ride_count = rural_city.groupby(["city"]).count()["ride_id"]
rural_avg_fare = rural_city.groupby(["city"]).mean()["fare"]
rural_driver_count = rural_city.groupby(["city"]).mean()["driver_count"]
```

```python
# Build the scatter plots for each city types
plt.scatter(urban_ride_count,urban_avg_fare,s=10*urban_driver_count,c='coral',edgecolor='black',linewidths=1,
             marker='o',alpha=0.5,label='Urban')


plt.scatter(sub_ride_count,sub_avg_fare,s=10*sub_driver_count,c='skyblue',edgecolor='black',linewidths=1,
             marker='o',alpha=0.5,label='Suburban')

plt.scatter(rural_ride_count,rural_avg_fare,s=10*rural_driver_count,c='gold',edgecolor='black',linewidths=1,
             marker='o',alpha=0.5,label='Rural')

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.xlim((0,50))
plt.grid(True)

# Create a legend
legend = plt.legend(fontsize="small", mode="Expanded",numpoints=1,scatterpoints=1,loc="best",
                    title="City Types",labelspacing=0.5)
legend.legendHandles[0]._sizes = [30]
legend.legendHandles[1]._sizes = [30]
legend.legendHandles[2]._sizes = [30]

# Incorporate a text label regarding circle size
plt.text(50, 30, "Note:\nCircle size correlates with driver count per city.")

# Save Figure
plt.savefig("../Resources/Pyber_analysis/Fig1.png")

# Show plot
plt.show()
```

```python
# Calculate Type Percents

type_percents = pyber_data.groupby(["type"]).sum()["fare"]

# Build Pie Chart

plt.pie(type_percents,labels=['Rural','Suburban','Urban'],colors=['gold','skyblue','coral'],explode=[0,0,0.1],
       autopct='%.2f%%',shadow=True,startangle=180,wedgeprops={'linewidth':.5,'edgecolor': 'black'})

plt.title('% Of Total Fares by Ride Type')

# Save Figure

plt.savefig("../Resources/Pyber_analysis/Fig2.png")

# Show Figure
plt.show()
```

```python
# Calculate Ride Percents

ride_percents = 100 * pyber_data.groupby(["type"]).count()["ride_id"] / pyber_data["ride_id"].count()

# Build Pie Chart

plt.pie(ride_percents,labels=['Rural','Suburban','Urban'],colors=['gold','skyblue','coral'],explode=[0,0,0.1],
       autopct='%.2f%%',shadow=True,startangle=180,wedgeprops={'linewidth':.5,'edgecolor': 'black'})

plt.title('% Of Total Rides by City Type')

# Save Figure

plt.savefig("../Resources/Pyber_analysis/Fig3.png")

# Show Figure
plt.show()
```

```python
# Calculate Driver Percents

driver_percents = city_df.groupby(["type"]).sum()["driver_count"] 

# Build Pie Chart

plt.pie(driver_percents,labels=['Rural','Suburban','Urban'],colors=['gold','skyblue','coral'],explode=[0,0,0.1],
       autopct='%.2f%%',shadow=True,startangle=180,wedgeprops={'linewidth':.5,'edgecolor': 'black'})

plt.title('% Of Total Drivers by City Type')

# Save Figure

plt.savefig("../Resources/Pyber_analysis/Fig4.png")

# Show Figure
plt.show()
```
